```python
import pandas as pd
import glob
```

```python
# Creates a DataFrame from the individual monthly bike rental csv files.
# I also added an 'id' column that will hopefully make any future joining easier.
df_list = []
bike_files = glob.glob('Bike_rental_data/data/JC-*-citibike-tripdata.csv')
for file in bike_files:
    file = pd.read_csv(file)
    df_list.append(file)

bike_df = pd.concat(df_list)
bike_df.reset_index(drop=True, inplace=True)
bike_df['id'] = bike_df.index
```

The citi bike System Data page states that the shortest a ride can be is 60 seconds and that they have already drop rides shorter than 60 seconds. 
I had to do some digging to find out how long a ride could be as it isn't stated in the System Data. Citi Bike considers that if a ride last longer than 24 hours the bike has been lost or stolen and changes a lost/stolen fee of $1,200. With this in mind I am going to add a 'valid_ride' column to the table. Rides lasting longer than 24 hours will be consider invalid rides rides between 60 seconds and 24 hours will be considered valid rides.
